```python
def fibonacci_loop(n):
    """
    Вычисляет n-ое число Фибоначчи с использованием цикла.
    Args:
        n: порядковый номер числа фибоначе
    Returns:
        n число Фибоначчи
    """
    
    a , b = 0 , 1
    for _ in range(n-2):
        a, b = b, a + b
    return a


fibonacci_loop(10)
```
