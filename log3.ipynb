```python
1. List
Описание: List представляет собой генерический тип данных, который используется для определения списка (массива) в Python. 
Пример использования: List[int] - список целых чисел.
2. Dict
Описание: Dict представляет собой генерический тип данных, который используется для определения словаря в Python. Словарь представляет собой структуру данных, состоящую из ключей и соответствующих им значений.
Пример использования: Dict[str, int] - словарь, в котором ключи являются строками, а значения - целыми числами.
  
3. Union
Описание: Union представляет собой генерический тип данных, который представляет собой объединение нескольких типов данных. Переменная или параметр функции, объявленные с использованием Union, могут принимать значения любого из указанных типов.
Пример использования: Union[int, float] - переменная, которая может быть либо целым числом, либо числом с плавающей точкой.
Модуль typing в Python является частью стандартной библиотеки и предоставляет возможность использования генерических типов данных для более точного определения типов в Python коде. Это улучшает читаемость кода, делает его более понятным и позволяет избежать ошибок при работе с типами данных.
Далее мы познакомились с магическими типами данных 
В Python существуют специальные "магические" методы (иногда их называют "магическими типами данных" или "магическими методами"), которые позволяют объектам вести себя как встроенные типы данных и управлять их поведением. Эти методы обладают двойными подчеркиваниями в начале и в конце имени, например __init__, __str__, __add__ и т. д.
Некоторые из наиболее часто используемых "магических" методов в Python включают:
1. __init__(): Конструктор объекта, вызывается при создании нового экземпляра класса.
2. __str__(): Метод, который возвращает строковое представление объекта, используемое функцией str().
3. __repr__(): Метод, который возвращает представление объекта в виде строки для вывода в интерпретаторе Python или при вызове функции repr().
4. __add__(), __sub__(), __mul__(), __truediv__(): Магические методы для перегрузки операторов сложения, вычитания, умножения и деления соответственно.
5. __len__(), __getitem__(), __setitem__(): Методы для перегрузки операторов длины объекта, доступа к элементам и присваивания элементам.
6. __eq__(), __ne__(), __lt__(), __gt__(), __le__(), __ge__(): Магические методы для перегрузки операторов сравнения.
Применение магических методов позволяет пользовательские классы вести себя так, как это ожидается от стандартных типов данных в Python, что позволяет более гибко управлять объектами и их поведением в различных контекстах.
```
